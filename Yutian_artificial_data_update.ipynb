For our dataset, most of them include two or more columns with numerical data, it can be used with linear regression. Linear regression can tell the relationships between multiple independent variables and a dependent variable. It allows us to understand how changes in one or more independent variables affect the dependent variable.

Sometimes, linear regression may not be a good way to describe data. Here, we are going to create a random dataset, let it tends to be an exponential form. Then I will choose different domain, to do the linear regression, and to see which domain fit the linear function better.

### Create data set 

```python
np.random.seed(1) 
# Generate random data with an exponential form in the range [0, 500]
x_range = [0, 500]
num_points = 1000

# Generate random x values within the specified range
x_data = np.random.uniform(x_range[0], x_range[1], num_points)

# Generate y values with an exponential form

y_data = np.exp(0.01 * x_data) + np.random.normal(0, 0.1, size=num_points)

# Select data points for linear approximation
x_fit_1 = x_data[(x_data >= 100) & (x_data <= 300)]
y_fit_1 = y_data[(x_data >= 100) & (x_data <= 300)]

x_fit_2 = x_data[(x_data >= 100) & (x_data <= 500)]
y_fit_2 = y_data[(x_data >= 100) & (x_data <= 500)]
```

```python
# Define linear function for fitting
def linear_func(x, m, c):
    return m * x + c

# Perform linear approximation for the first range [100,125]
params_1, _ = curve_fit(linear_func, x_fit_1, y_fit_1)

# Perform linear approximation for the second range [100,200]
params_2, _ = curve_fit(linear_func, x_fit_2, y_fit_2)
```

```python
# Plot the original data and linear approximations
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Original Data', color='blue', alpha=0.5)
plt.plot(x_fit_1, linear_func(x_fit_1, *params_1), label='Linear Fit [100, 300]', color='red')
plt.plot(x_fit_2, linear_func(x_fit_2, *params_2), label='Linear Fit [100, 500]', color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Approximation of Exponential Data')
plt.legend()
plt.grid(True)
plt.show()

print("Linear Fit Parameters for [100, 300]: m =", params_1[0], ", c =", params_1[1])
print("Linear Fit Parameters for [100, 500]: m =", params_2[0], ", c =", params_2[1])
```

### Examine the statistics

From the above graph, we can see that for a smaller domain, linear regression fits better for the exponential data, but when the domain becomes large, the linear regression did not look good. Now, let's see the statistical information.

```python
# Fit OLS regression for the first range [100, 300]
x_fit_1_ols = sm.add_constant(x_fit_1)
model_1 = sm.OLS(y_fit_1, x_fit_1_ols)
results_1 = model_1.fit()

# Fit OLS regression for the second range [100, 500]
x_fit_2_ols = sm.add_constant(x_fit_2)
model_2 = sm.OLS(y_fit_2, x_fit_2_ols)
results_2 = model_2.fit()

# Print OLS regression results
print("OLS Regression Results for [100, 300]:")
print(results_1.summary())
print("\nOLS Regression Results for [100, 500]:")
print(results_2.summary())
```

From the two outputs, we have the following results:

Compared the two R-squared, the smaller domain is larger than the bigger domain, which means more variance can be explained better by smaller domain.

F-statistic is the overall F-statistic for the regression model.

Probs are both less than 0.05, it tells us whether or not the regression model as a whole is statistically significant. Our p values are less than 0.05, which means $x$ did have statistically significant association with $y$.
